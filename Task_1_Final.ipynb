```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('/content/CarPrice_Assignment.csv')
cars.head()
```

```python
CompanyName = cars['CarName'].apply(lambda x : x.split(' ')[0])
cars.insert(3,"CompanyName",CompanyName)
cars.drop(['CarName'],axis=1,inplace=True)
cars.head()
```

```python
cars.CompanyName.unique()
```

```python
cars.CompanyName = cars.CompanyName.str.lower()

def replace_name(a,b):
    cars.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

cars.CompanyName.unique()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=cars.price)

plt.show()
print(cars.price.describe(percentiles = [0.25,0.50,0.75,0.85,0.90,1]))
```

```python
plt.figure(figsize=(25, 6))

plt.subplot(1,3,1)
plt1 = cars.CompanyName.value_counts().plot(kind ='bar')
plt.title('Companies Histogram')
plt1.set(xlabel = 'Car company', ylabel='Frequency of company')

plt.subplot(1,3,2)
plt1 = cars.fueltype.value_counts().plot(kind = 'bar')
plt.title('Fuel Type Histogram')
plt1.set(xlabel = 'Fuel Type', ylabel='Frequency of fuel type')

plt.subplot(1,3,3)
plt1 = cars.carbody.value_counts().plot(kind = 'bar')
plt.title('Car Type Histogram')
plt1.set(xlabel = 'Car Type', ylabel='Frequency of Car type')

plt.show()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,2)
plt.title('Symboling vs Price')
sns.boxplot(x=cars.symboling, y=cars.price, palette=("cubehelix"))

plt.show()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,2)
plt.title('Engine Type vs Price')
sns.boxplot(x=cars.enginetype, y=cars.price, palette=("PuBuGn"))

plt.show()

cars_2 = pd.DataFrame(cars.groupby(['enginetype'])['price'].mean().sort_values(ascending = False))
cars_2.plot.bar(figsize=(8,6))
plt.title('Engine Type vs Average Price')
plt.show()
```

```python
def pp(x,y,z):
    sns.pairplot(cars, x_vars=[x,y,z], y_vars='price',height=4, aspect=1, kind='scatter')
    plt.show()

pp('enginesize', 'boreratio', 'stroke')
pp('compressionratio', 'horsepower', 'peakrpm')
pp('wheelbase', 'citympg', 'highwaympg')
```

```python
plt.figure(figsize=(8,6))

plt.title('Fuel system vs Price')
sns.scatterplot(x=cars['fuelsystem'],y=cars['price'],hue=cars['drivewheel'])
plt.xlabel('Fuel Economy')
plt.ylabel('Price')

plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(25, 6))

cars_2 = pd.DataFrame(cars.groupby(['fuelsystem','drivewheel','aspiration'])['price'].mean().unstack(fill_value=0))
cars_2.plot.bar()
plt.title('Car Range vs Average Price')
plt.show()
```

```python
df_modified=cars[['symboling','fueltype','aspiration','doornumber','drivewheel','enginelocation','wheelbase','carlength','carwidth','carheight','curbweight','enginetype','cylindernumber','enginesize','fuelsystem','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price']]
df_modified.head()
```

```python
df_modified=pd.get_dummies(df_modified,drop_first=True)
df_modified.head()
```

```python
from sklearn.model_selection import train_test_split
x = df_modified
y = cars['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,test_size=0.8,random_state=0)
```
